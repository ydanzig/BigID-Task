## A script that finds the location a given names in a large text file
## results will save in a dicionary of tuples structure.

import os
import urllib.request

```python
#Function returns a list of indexes of the inner position of substring inside the string.
#returns an empty list is nothing was found.
#case sensitive parameter is default False. 
def find_all(string,substring,caseSensitive = False):
    length = len(substring)          #Length of the substring
    p = 0                            #Position running index
    indexes = []                     #Saving results in a list
    while p < (len(string)-length+1):
        if caseSensitive:
            if string[p:p+length] == substring:
                indexes.append(p)
        else:
            if string[p:p+length].lower() == substring.lower():
                indexes.append(p)
        p+=1
    return indexes
```

```python
#Function takes a URL (text file) and list of words and returns a dictionary that contains the occurence 
#of the row inner character place that each word is located in the text file.
def matcher(url,Names):
    file = urllib.request.urlopen(url)
    #Initiating the result dictionary. it will contain search results.
    result = {}
    for i in Names:
        result[i] = []
    lineCounter = 0             #Line Counter in the text file          
    for line in file:
        for name in Names:
            decoded_line = line.decode("utf-8")
            lineResult = find_all(decoded_line,name)
            for e in lineResult:
                result[name].append((lineCounter,e))
        lineCounter+=1
    return result
```

```python
#Names to search in the url below: 
url =  'http://norvig.com/big.txt'
Names = ['James','John','Robert','Michael','William','David','Richard','Charles','Joseph','Thomas',
         'Christopher','Daniel','Paul','Mark','Donald','George','Kenneth','Steven', 'Edward',
        'Brian','Ronald','Anthony','Kevin','Jason','Matthew', 'Gary', 'Timothy','Jose','Larry',
        'Jeffrey','Frank','Scott','Eric','Stephen','Andrew','Raymond','Gregory','Joshua','Jerry',
         'Dennis', 'Walter','Patrick', 'Peter', 'Harold','Douglas', 'Henry', 'Carl','Arthur','Ryan','Roger']
```

```python
#Exacuting the script:
result = matcher(url,Names)
```
